7. Split the dataset into a training set and a testing set. Fit a linear regression model to the training set using the "LSTAT" feature as the independent variable and "MEDV" as the dependent variable. Evaluate the model on the testing set using the mean squared error (MSE) and R-squared (R2) metrics. 

```python
#Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(bostondata[["lstat"]], bostondata["medv"], test_size=0.4, random_state=44)

#Creating an instance of the Linear Regression Class
reg = LinearRegression()

#Fitting the model onto the training data
reg.fit(X_train, y_train)

#Evaluating the model on the testing data
y_pred = reg.predict(X_test)

#Evaluating the performance of the model on the testing data
mse = mean_squared_error(y_test, y_pred)
rS = r2_score(y_test,y_pred)

print("Mean Squared Error: ", mse)
print("R-squared: ", rS)
```

10. Preprocess the dataset by converting the "MEDV" variable to a binary variable ("MEDV_binary") based on whether its value is above or below the median value, and splitting the dataset into a training set and a testing set. 

```python
#Converting "MEDV" to a binary variable based on if the value is above or below the median value
bostondata["medv_binary"] = np.where(bostondata["medv"] >= bostondata["medv"].median(), 1, 0)

# Split the preprocessed data into training and testing sets
X_train4, X_test4, y_train4, y_test4 = train_test_split(bostondata.drop(["medv", "medv_binary"], axis=1), bostondata["medv_binary"], test_size=0.6, random_state=47)
```

11. Fit a logistic regression model to the dataset using all the features except "MEDV" as the independent variables and "MEDV_binary" as the dependent variable. Print the coefficients of the model.

```python
#Creating an instance of the Logisitic Regression Class
logreg = LogisticRegression(max_iter=5000)

# Fit a Logisitic regression model to the training data
logreg.fit(X_train4, y_train4)

# Print the coefficients of the model
print("Coefficients:", logreg.coef_)
```

12. Evaluate the model from question 11 on the testing set using the accuracy and F1-score metrics.

```python
# make predictions on the testing set
y_pred = logreg.predict(X_test4)

# calculate accuracy and F1-score
acc = accuracy_score(y_test4, y_pred)
f1 = f1_score(y_test4, y_pred)

print("Accuracy:", acc)
print("F1-score:", f1)
```
